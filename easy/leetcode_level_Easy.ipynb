# Two Sum

URL: https://leetcode.com/problems/two-sum/

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        dict = {}
        
        print('len(nums): ',len(nums))
        for i in range(len(nums)):
            print('i: ',i)
            x = nums[i]
            print('x: ',x)
            if target-x in dict:
                print('target-x: ',target-x)
                return [dict[target-x], i]
            dict[x] = i
            print('dict[x]: ',dict[x])
            print('dict: ',dict)
            
# Create an instance of the Solution class
solution = Solution()

# Call the twoSum method with your input values
nums = [2, 11, 9, 7]
target = 9
result = solution.twoSum(nums, target)

# Print the result
print(result)
```

# Palindrome Number

URL: https://leetcode.com/problems/palindrome-number/

```python
class Solution:
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0:
            print('x: ',x)
            return False
        rev = 0
        copy = x
        while copy != 0:
            print('copy(inside while loop): ', copy)
            rev = rev*10 + copy%10
            print('rev: ',rev)
            copy = copy//10
            print('copy: ', copy)
        if rev == x:
            return True
        else:
            return False
        
solution = Solution()

result = solution.isPalindrome(1221)

# Print the result
print(result)
```

# Roman to Integer

URL: https://leetcode.com/problems/roman-to-integer/

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        roman_dict = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        total = 0
        prev_value = 0

        for symbol in s[::-1]: # Start from the end of the string
            print('symbol: ',symbol)
            value = roman_dict[symbol]
            print('value: ',value)
            if value < prev_value:
                total -= value
                print('if total: ', total)
            else:
                total += value
                print('else total: ', total)
            prev_value = value
            print('prev_value: ',prev_value)
        return total

solution = Solution()
result = solution.romanToInt("MCXII")
print(result)
```

# Longest Common Prefix

URL:https://leetcode.com/problems/longest-common-prefix/

```python
class Solution:
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        # To handle the case when the input list strs is empty.
        if not strs:
            return ""
        
        # Sort the input strings to easily compare the first and last strings
        strs.sort()
        print('strs: ',strs)
        
        # Find the common prefix between the first and last strings
        common_prefix = []
        for i in range(len(strs[0])):
            print('len(strs[0]): ',len(strs[0]))
            print('i: ',i)
            if i < len(strs[-1]) and strs[0][i] == strs[-1][i]:
                print('len(strs[-1]): ',len(strs[-1]))
                print('strs[-1][i]: ',strs[-1][i])
                common_prefix.append(strs[0][i])
                print('common_prefix: ',common_prefix)
            else:
                break
        
        return "".join(common_prefix)

strs1 = ["flower","flow","flight"]
solution = Solution()
result = solution.longestCommonPrefix(strs1)
print(result)
```

```python
class Solution:
    def isValid(self, s):
        if not s:
            return False
        else:
            stack = []
            balanced = True
            index = 0
            while index < len(s) and balanced:
                print('len(s): ',len(s))
                symbol = s[index]
                print('s[index]: ',s[index])
                if symbol in "({[":
                    print("-if-")
                    print('symbol: ',symbol)
                    stack.append(symbol)
                    print('stack: ',stack)
                else:
                    print("-else-")
                    if not stack:
                        print("--if not stack--")
                        balanced = False
                    else:
                        print("--else--")
                        print('stack: ',stack)
                        top = stack.pop()
                        print('top: ',top)
                        print('stack: ',stack)
                        if not self.matches(top, symbol):
                            balanced = False
                index = index + 1

            if balanced and not stack:
                return True
            else:
                return False

    def matches(self, open, close):
        openings = "({["
        closings = ")}]"
        return openings.index(open) == closings.index(close)


# Create an instance of the Solution class
solution = Solution()

# Test cases
test_cases = ["()", "()[]{}", "(]", "{[()]}", "[(])", "{[(]})"]

for s in test_cases:
    is_valid = solution.isValid(s)
    print(f"Input: {s}, Output: {is_valid}",'\n')
```

# Merge Two Sorted Lists

URL: https://leetcode.com/problems/merge-two-sorted-lists/

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None
        
class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        if l1 == None and l2 == None:
            return None
        elif l1 != None and l2 == None:
            return l1
        elif l2 != None and l1 == None:
            return l2
        else:
            dummy = ListNode(0)
            print('dummy: ',dummy,'\n')
            p = dummy
            print('p: ',p,'\n')
            
            while l1 != None and l2 != None:
                print('l1.val: ',l1.val,'  &  l2.val: ',l2.val,'\n\n')
                if l1.val < l2.val:
                    print('while l1 != None and l2 != None IF:       ')
                    print('l1.val: ',l1.val,'  &  l2.val: ',l2.val,'\n')
                    p.next = l1
                    l1 = l1.next
                else:
                    print('while l1 != None and l2 != None ELSE:       ')
                    p.next = l2
                    l2 = l2.next
                p = p.next
                print('-'*50,'\n','while p: ',p,'\n')
                
            if l1 != None:
                p.next = l1
            if l2 != None:
                p.next = l2
            
            return dummy.next


# Create an instance of the Solution class
solution = Solution()

#Test Case for the input
list1 = [1,2,4] 
print('list1: ',list1,'\n')

list2 = [1,3,4]
print('list2: ',list2,'\n')

# Helper function to convert list to linked list
def list_to_linkedlist(lst):
    if not lst:
        return None
    head = ListNode(lst[0])
    current = head
    for val in lst[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

# Convert the input lists to linked lists
linkedlist1 = list_to_linkedlist(list1)
linkedlist2 = list_to_linkedlist(list2)

# Call the mergeTwoLists method
result_linkedlist = solution.mergeTwoLists(linkedlist1, linkedlist2)

# Helper function to convert linked list to list
def linkedlist_to_list(node):
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result

# Convert the result linked list to a list and print
result_list = linkedlist_to_list(result_linkedlist)
print(result_list)
```

# Remove Duplicates from Sorted Array

URL: https://leetcode.com/problems/remove-duplicates-from-sorted-array/

```python
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums) < 2:
            print('len(nums)',len(nums),'\n')
            return len(nums)
        else:
            j = 0
            i = 1
            while i < len(nums):
                print('i',i,'\n')
                if nums[j] == nums[i]:
                    print('nums[j]',nums[j],' & ','nums[i]',nums[i])
                    i += 1
                else:
                    print('nums[j]',nums[j],' & ','nums[i]',nums[i])
                    j += 1
                    nums[j] = nums[i]
                    i += 1
            return j+1

        
solution_instance = Solution()
nums = [0,0,1,1,1,2,2,3,3,4]
result_length = solution_instance.removeDuplicates(nums)

print(f"Original List: {nums}")
print(f"Length of List after removing duplicates: {result_length}")
print(f"List after removing duplicates: {nums[:result_length]}")
```

# Remove Element

URL: https://leetcode.com/problems/remove-element/

```python
class Solution(object):
    def removeElement(self, nums, val):
        """
        :type nums: List[int]
        :type val: int
        :rtype: int
        """
        if val == []:
            print('val: ',val,'\n')
            return 0
        else:
            i = 0
            j = 0
            while j < len(nums):
                print('j: ',j,'   len(nums): ',len(nums),'\n')
                if nums[j] == val:
                    j += 1
                else:
                    print('nums[i]: ',nums[i],'    nums[j]: ', nums[j],'\n')
                    nums[i] = nums[j]
                    i += 1
                    j += 1
            return len(nums[0:i])

# Example usage
sol = Solution()

# Test case 1
nums1 = [3, 2, 2, 3]
val1 = 2
result1 = sol.removeElement(nums1, val1)
print(f"Result 1: {result1}, Modified nums1: {nums1}")

# Test case 2
nums2 = [0, 1, 2, 2, 3, 0, 4, 2]
val2 = 2
result2 = sol.removeElement(nums2, val2)
print(f"Result 2: {result2}, Modified nums2: {nums2}")
```

 # Find the Index of the First Occurrence in a String
    
URL : https://leetcode.com/problems/find-the-index-of-the-first-occurrence-in-a-string

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if not needle:
            return 0
        # Iterate through the haystack string
        for i in range(len(haystack) - len(needle) + 1):
            print('i: ',i,'\n')
            # Check if substring starting at index i matches the needle
            if haystack[i:i + len(needle)] == needle:
                print('needle: ',needle,'\n')
                return i

        # If no match is found, return -1
        return -1

# Example usage:
solution = Solution()
print(solution.strStr("sadbutsad", "sad"))

# print(solution.strStr("leetcode", "leeto"))
```

# Search Insert Position

URL : https://leetcode.com/problems/search-insert-position/

```python
from typing import List

class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        
        left, right = 0, len(nums) - 1
        print('left: ',left,'     right: ',right)

        while left <= right:
            mid = left + (right - left) // 2
            print('mid: ',mid)

            if nums[mid] == target:
                print("IF\ntarget: ",target)
                return mid
            elif nums[mid] < target:
                print("ELIF\ntarget: ",target)
                left = mid + 1
            else:
                print("ELSE\ntarget: ",target)
                right = mid - 1

        # If the target is not found, 'left' represents the index where it should be inserted
        return left

# Example usage:
solution = Solution()

nums1, target1 = [1, 3, 5, 6], 5
result1 = solution.searchInsert(nums1, target1)
print(result1)

nums2, target2 = [1, 3, 5, 6], 2
result2 = solution.searchInsert(nums2, target2)
print(result2)

nums3, target3 = [1, 3, 5, 6], 7
result3 = solution.searchInsert(nums3, target3)
print(result3)
```

# Plus one

URL: https://leetcode.com/problems/plus-one/

```python
from typing import List

class Solution(object):
    def plusOne(self, digits: List[int]) -> List[int]:
        if len(digits) == 0:
            print('len(digits): ',len(digits),'\n')
            return [0]

        carry = 1
        new_digits = []

        for digit in reversed(digits):
            print('digit: ',digit,'\n')
            running_sum = digit + carry
            new_digits.append(running_sum % 10)
            carry = running_sum // 10
            print('carry: ',carry,'\n')

        if carry == 1:
            new_digits.append(1)

        return new_digits[::-1]

# Testing the updated code
s = Solution()
result = s.plusOne([1, 2, 3])
print(result)

result = s.plusOne([4, 3, 2, 1])
print(result)

result = s.plusOne([9])
print(result)
```

# Add Binary

URL : https://leetcode.com/problems/add-binary/

```python
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        result = []
        carry = 0
        
        # Traverse the strings from right to left
        i, j = len(a) - 1, len(b) - 1
        print('i: ',i)
        print('j: ',j)
        
        while i >= 0 or j >= 0 or carry:
            # Extract the digits from the strings
            digit_a = int(a[i]) if i >= 0 else 0
            digit_b = int(b[j]) if j >= 0 else 0
            
            # Calculate the sum and carry
            total_sum = digit_a + digit_b + carry
            carry = total_sum // 2
            current_digit = total_sum % 2
            
            # Insert the current digit at the beginning of the result list
            result.insert(0, str(current_digit))
            
            # Move to the next digits in both strings
            i -= 1
            j -= 1
        
        # Convert the result list to a string
        return ''.join(result)

# Example usage:
solution = Solution()
print(solution.addBinary("11", "1"))  # Output: "100"
print(solution.addBinary("1010", "1011"))  # Output: "10101"
```

# Remove Duplicates from Sorted List
URL : https://leetcode.com/problems/remove-duplicates-from-sorted-list/description/

```python
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution(object):
    def deleteDuplicates(self, head):
        if head is None:
            return head
        elif head.next is None:
            return head
        else:
            lookup = {}
            current = head
            prev = head
            while current is not None:
                if current.val in lookup:
                    prev.next = current.next
                else:
                    lookup[current.val] = True
                    prev = current
                current = current.next
            return head

# Helper function to convert a list to a linked list
def list_to_linked_list(lst):
    if not lst:
        return None
    head = ListNode(lst[0])
    current = head
    for val in lst[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

# Helper function to convert a linked list to a list
def linked_list_to_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

# Test case
input_list = [1, 1, 2, 3, 3]
head = list_to_linked_list(input_list)

solution = Solution()
new_head = solution.deleteDuplicates(head)

# Validate the result
result_list = linked_list_to_list(new_head)
print(result_list)
```

# Merge Sorted Array
URL : https://leetcode.com/problems/merge-sorted-array/

```python
from typing import List

class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        # Initialize pointers for nums1 and nums2
        i, j, k = m-1, n-1, m+n-1

        # Merge nums1 and nums2 from the end
        while i >= 0 and j >= 0:
            if nums1[i] > nums2[j]:
                nums1[k] = nums1[i]
                i -= 1
            else:
                nums1[k] = nums2[j]
                j -= 1
            k -= 1

        # If there are remaining elements in nums2, copy them to nums1
        while j >= 0:
            nums1[k] = nums2[j]
            j -= 1
            k -= 1

# Example usage:
# Initialize nums1, m, nums2, and n
nums1 = [1, 2, 3, 0, 0, 0]
m = 3
nums2 = [2, 5, 6]
n = 3

# Create an instance of the Solution class
solution = Solution()
# Call the merge function
solution.merge(nums1, m, nums2, n)
# Print the result
print(nums1)  # Output: [1,2,2,3,5,6]
```

# Binary Tree Inorder Traversal
URL : https://leetcode.com/problems/binary-tree-inorder-traversal

```python
from typing import Optional, List

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        result = []
        stack = []
        current = root

        while current or stack:
            # Traverse left subtree and push nodes onto the stack
            while current:
                stack.append(current)
                current = current.left

            # Visit the current node
            current = stack.pop()
            result.append(current.val)

            # Move to the right subtree
            current = current.right

        return result

# Example usage
# Example 1: [1, null, 2, 3]
root1 = TreeNode(1)
root1.right = TreeNode(2)
root1.right.left = TreeNode(3)

solution = Solution()
result1 = solution.inorderTraversal(root1)
print(result1)

# Example 2: Empty tree
root2 = None
result2 = solution.inorderTraversal(root2)
print(result2)

# Example 3: Single-node tree [1]
root3 = TreeNode(1)
result3 = solution.inorderTraversal(root3)
print(result3)
```

# Same Tree
URL : https://leetcode.com/problems/same-tree

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def isSameTree(self, p, q):
        if p is None and q is None:
            return True
        elif p is None or q is None:
            return False
        else:
            return p.val == q.val and self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)

# Example 1
p1 = TreeNode(1)
p1.left = TreeNode(2)
p1.right = TreeNode(3)

q1 = TreeNode(1)
q1.left = TreeNode(2)
q1.right = TreeNode(3)

sol = Solution()
print(sol.isSameTree(p1, q1))  

# Example 2
p2 = TreeNode(1)
p2.left = TreeNode(2)

q2 = TreeNode(1)
q2.right = TreeNode(2)

print(sol.isSameTree(p2, q2))  

# Example 3
p3 = TreeNode(1)
p3.left = TreeNode(2)
p3.right = TreeNode(1)

q3 = TreeNode(1)
q3.left = TreeNode(1)
q3.right = TreeNode(2)

print(sol.isSameTree(p3, q3))
```
