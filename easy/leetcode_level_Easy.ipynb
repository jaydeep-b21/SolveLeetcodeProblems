# Two Sum

URL: https://leetcode.com/problems/two-sum/

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        dict = {}
        
        print('len(nums): ',len(nums))
        for i in range(len(nums)):
            print('i: ',i)
            x = nums[i]
            print('x: ',x)
            if target-x in dict:
                print('target-x: ',target-x)
                return [dict[target-x], i]
            dict[x] = i
            print('dict[x]: ',dict[x])
            print('dict: ',dict)
            
# Create an instance of the Solution class
solution = Solution()

# Call the twoSum method with your input values
nums = [2, 11, 9, 7]
target = 9
result = solution.twoSum(nums, target)

# Print the result
print(result)
```

# Palindrome Number

URL: https://leetcode.com/problems/palindrome-number/

```python
class Solution:
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0:
            print('x: ',x)
            return False
        rev = 0
        copy = x
        while copy != 0:
            print('copy(inside while loop): ', copy)
            rev = rev*10 + copy%10
            print('rev: ',rev)
            copy = copy//10
            print('copy: ', copy)
        if rev == x:
            return True
        else:
            return False
        
solution = Solution()

result = solution.isPalindrome(1221)

# Print the result
print(result)
```

```python
# Method 2
class Solution:
    def isPalindrome(self, x):
        if x < 0:
            return False
        rev = 0
        copy = x
        while copy != 0:
            rev = rev * 10 + copy % 10
            copy = copy // 10
        return rev == x

solution = Solution()
result = solution.isPalindrome(12021)
print(result)
```

```python
# Method 3 (Palindrome for strings)
class Solution:
    def isPalindrome(self, x: int) -> bool:
        # Convert the integer to a string
        x_str = str(x)

        # Check if the string is equal to its reverse
        return x_str == x_str[::-1]

solution = Solution()
result = solution.isPalindrome("12r21")
print(result)
```

# Roman to Integer

URL: https://leetcode.com/problems/roman-to-integer/

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        roman_dict = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        total = 0
        prev_value = 0

        for symbol in s[::-1]: # Start from the end of the string
            print('symbol: ',symbol)
            value = roman_dict[symbol]
            print('value: ',value)
            if value < prev_value:
                total -= value
                print('if total: ', total)
            else:
                total += value
                print('else total: ', total)
            prev_value = value
            print('prev_value: ',prev_value)
        return total

solution = Solution()
result = solution.romanToInt("MCXII")
print(result)
```

# Longest Common Prefix

URL:https://leetcode.com/problems/longest-common-prefix/

```python
class Solution:
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        # To handle the case when the input list strs is empty.
        if not strs:
            return ""
        
        # Sort the input strings to easily compare the first and last strings
        strs.sort()
        print('strs: ',strs)
        
        # Find the common prefix between the first and last strings
        common_prefix = []
        for i in range(len(strs[0])):
            print('len(strs[0]): ',len(strs[0]))
            print('i: ',i)
            if i < len(strs[-1]) and strs[0][i] == strs[-1][i]:
                print('len(strs[-1]): ',len(strs[-1]))
                print('strs[-1][i]: ',strs[-1][i])
                common_prefix.append(strs[0][i])
                print('common_prefix: ',common_prefix)
            else:
                break
        
        return "".join(common_prefix)

strs1 = ["flower","flow","flight"]
solution = Solution()
result = solution.longestCommonPrefix(strs1)
print(result)
```

# Valid Parentheses

URL: https://leetcode.com/problems/valid-parentheses/

```python
class Solution:
    def isValid(self, s):
        if not s:
            return False
        else:
            stack = []
            balanced = True
            index = 0
            while index < len(s) and balanced:
                print('len(s): ',len(s))
                symbol = s[index]
                print('s[index]: ',s[index])
                if symbol in "({[":
                    print("-if-")
                    print('symbol: ',symbol)
                    stack.append(symbol)
                    print('stack: ',stack)
                else:
                    print("-else-")
                    if not stack:
                        print("--if not stack--")
                        balanced = False
                    else:
                        print("--else--")
                        print('stack: ',stack)
                        top = stack.pop()
                        print('top: ',top)
                        print('stack: ',stack)
                        if not self.matches(top, symbol):
                            balanced = False
                index = index + 1

            if balanced and not stack:
                return True
            else:
                return False

    def matches(self, open, close):
        openings = "({["
        closings = ")}]"
        return openings.index(open) == closings.index(close)


# Create an instance of the Solution class
solution = Solution()

# Test cases
test_cases = ["()", "()[]{}", "(]", "{[()]}", "[(])", "{[(]})"]

for s in test_cases:
    is_valid = solution.isValid(s)
    print(f"Input: {s}, Output: {is_valid}",'\n')
```

# Merge Two Sorted Lists

URL: https://leetcode.com/problems/merge-two-sorted-lists/

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None
        
class Solution(object):
    def mergeTwoLists(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        if l1 == None and l2 == None:
            return None
        elif l1 != None and l2 == None:
            return l1
        elif l2 != None and l1 == None:
            return l2
        else:
            dummy = ListNode(0)
            print('dummy: ',dummy,'\n')
            p = dummy
            print('p: ',p,'\n')
            
            while l1 != None and l2 != None:
                print('l1.val: ',l1.val,'  &  l2.val: ',l2.val,'\n\n')
                if l1.val < l2.val:
                    print('while l1 != None and l2 != None IF:       ')
                    print('l1.val: ',l1.val,'  &  l2.val: ',l2.val,'\n')
                    p.next = l1
                    l1 = l1.next
                else:
                    print('while l1 != None and l2 != None ELSE:       ')
                    p.next = l2
                    l2 = l2.next
                p = p.next
                print('-'*50,'\n','while p: ',p,'\n')
                
            if l1 != None:
                p.next = l1
            if l2 != None:
                p.next = l2
            
            return dummy.next


# Create an instance of the Solution class
solution = Solution()

#Test Case for the input
list1 = [1,2,4] 
print('list1: ',list1,'\n')

list2 = [1,3,4]
print('list2: ',list2,'\n')

# Helper function to convert list to linked list
def list_to_linkedlist(lst):
    if not lst:
        return None
    head = ListNode(lst[0])
    current = head
    for val in lst[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

# Convert the input lists to linked lists
linkedlist1 = list_to_linkedlist(list1)
linkedlist2 = list_to_linkedlist(list2)

# Call the mergeTwoLists method
result_linkedlist = solution.mergeTwoLists(linkedlist1, linkedlist2)

# Helper function to convert linked list to list
def linkedlist_to_list(node):
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result

# Convert the result linked list to a list and print
result_list = linkedlist_to_list(result_linkedlist)
print(result_list)
```

# Remove Duplicates from Sorted Array

URL: https://leetcode.com/problems/remove-duplicates-from-sorted-array/

```python
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if len(nums) < 2:
            print('len(nums)',len(nums),'\n')
            return len(nums)
        else:
            j = 0
            i = 1
            while i < len(nums):
                print('i',i,'\n')
                if nums[j] == nums[i]:
                    print('nums[j]',nums[j],' & ','nums[i]',nums[i])
                    i += 1
                else:
                    print('nums[j]',nums[j],' & ','nums[i]',nums[i])
                    j += 1
                    nums[j] = nums[i]
                    i += 1
            return j+1

        
solution_instance = Solution()
nums = [0,0,1,1,1,2,2,3,3,4]
result_length = solution_instance.removeDuplicates(nums)

print(f"Original List: {nums}")
print(f"Length of List after removing duplicates: {result_length}")
print(f"List after removing duplicates: {nums[:result_length]}")
```

# Remove Element

URL: https://leetcode.com/problems/remove-element/

```python
class Solution(object):
    def removeElement(self, nums, val):
        """
        :type nums: List[int]
        :type val: int
        :rtype: int
        """
        if val == []:
            print('val: ',val,'\n')
            return 0
        else:
            i = 0
            j = 0
            while j < len(nums):
                print('j: ',j,'   len(nums): ',len(nums),'\n')
                if nums[j] == val:
                    j += 1
                else:
                    print('nums[i]: ',nums[i],'    nums[j]: ', nums[j],'\n')
                    nums[i] = nums[j]
                    i += 1
                    j += 1
            return len(nums[0:i])

# Example usage
sol = Solution()

# Test case 1
nums1 = [3, 2, 2, 3]
val1 = 2
result1 = sol.removeElement(nums1, val1)
print(f"Result 1: {result1}, Modified nums1: {nums1}")

# Test case 2
nums2 = [0, 1, 2, 2, 3, 0, 4, 2]
val2 = 2
result2 = sol.removeElement(nums2, val2)
print(f"Result 2: {result2}, Modified nums2: {nums2}")
```
